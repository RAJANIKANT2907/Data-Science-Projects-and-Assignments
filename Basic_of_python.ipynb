### Exercise 1: Prime Numbers

This program defines a function `is_prime` that checks if a given number is prime. It iterates from 2 up to the square root of the number to find any divisors. If no divisors are found, the number is prime. The program then takes user input, converts it to an integer, and uses the `is_prime` function to determine and print whether the number is prime.

```python
import math

def is_prime(num):
    if num <= 1:
        return False
    for i in range(2, int(math.sqrt(num)) + 1):
        if num % i == 0:
            return False
    return True

# Get input from the user
number_to_check = int(input("Enter a number to check if it's prime: "))

if is_prime(number_to_check):
    print(f"{number_to_check} is a prime number.")
else:
    print(f"{number_to_check} is not a prime number.")
```

### Exercise 2: Product of Random Numbers

This program generates two random integers between 1 and 10. It then prompts the user to enter the product of these two numbers. Finally, it checks if the user's answer matches the actual product and provides feedback.

```python
import random

# Generate two random numbers
num1 = random.randint(1, 10)
num2 = random.randint(1, 10)

print(f"What is the product of {num1} and {num2}?")

# Get user's answer
try:
    user_answer = int(input("Your answer: "))
except ValueError:
    print("Invalid input. Please enter an integer.")
else:
    correct_answer = num1 * num2
    if user_answer == correct_answer:
        print("Correct! Well done!")
    else:
        print(f"Incorrect. The correct answer is {correct_answer}.")
```

### Exercise 3: Squares of Even/Odd Numbers

This script iterates through numbers in the range of 100 to 200 (inclusive). It has been configured to process **even numbers**. For each even number, it calculates and prints its square.

```python
# Choice: Process even numbers
# To process odd numbers, change 'num % 2 == 0' to 'num % 2 != 0'

print("Squares of even numbers between 100 and 200:")
for num in range(100, 201):  # Range 100 to 200 inclusive
    if num % 2 == 0:  # Check if the number is even
        square = num ** 2
        print(f"The square of {num} is {square}")
```

### Exercise 4: Hospital Billing System

This program calculates the final bill for a hospital patient based on the provided charges and discount rules. It defines a function `calculate_hospital_bill` that takes the number of days stayed, lab tests, and medicine charges as input. It then calculates the total bill and applies a 10% discount if the total exceeds ₹10,000. Finally, it calls this function with the specific patient details given in the problem statement.

```python
def calculate_hospital_bill(days_stayed, num_lab_tests, medicine_charges):
    # Room charges per day
    room_charge_per_day = 2000
    total_room_charges = days_stayed * room_charge_per_day

    # Doctor consultation fee (flat)
    doctor_consultation_fee = 1500

    # Lab test charges
    lab_test_cost_per_test = 300
    total_lab_charges = num_lab_tests * lab_test_cost_per_test

    # Medicine charges
    total_medicine_charges = medicine_charges

    # Calculate total bill before discount
    total_bill_before_discount = (
        total_room_charges
        + doctor_consultation_fee
        + total_lab_charges
        + total_medicine_charges
    )

    # Apply discount if applicable
    discount_percentage = 0.10  # 10%
    if total_bill_before_discount > 10000:
        discount_amount = total_bill_before_discount * discount_percentage
        final_bill = total_bill_before_discount - discount_amount
        print(f"Total bill before discount: ₹{total_bill_before_discount:,.2f}")
        print(f"Discount applied (10%): ₹{discount_amount:,.2f}")
    else:
        final_bill = total_bill_before_discount
        print(f"Total bill: ₹{total_bill_before_discount:,.2f} (No discount applied)")

    return final_bill

# Patient details from the problem statement
days_stayed_patient = 4
num_lab_tests_patient = 3
medicine_charges_patient = 2400

# Calculate and display the final bill
print("\nHospital Bill Calculation:")
final_patient_bill = calculate_hospital_bill(
    days_stayed_patient,
    num_lab_tests_patient,
    medicine_charges_patient
)

print(f"Final bill for the patient: ₹{final_patient_bill:,.2f}")
```
