```python
import pandas as pd

# Load the dataset
anime_df = pd.read_csv('/content/anime.csv')

# Display the first 5 rows of the DataFrame
display(anime_df.head())
```

```python
# Display the DataFrame's information to check data types and non-null counts
display(anime_df.info())
```

```python
# Display descriptive statistics for numerical columns
display(anime_df.describe())
```

```python
# Handle missing values

# For 'genre' and 'type', fill missing values with 'Unknown'
anime_df['genre'] = anime_df['genre'].fillna('Unknown')
anime_df['type'] = anime_df['type'].fillna('Unknown')

# For 'rating', fill missing values with the mean rating
anime_df['rating'] = anime_df['rating'].fillna(anime_df['rating'].mean())

# Convert 'episodes' to numeric. First, replace 'Unknown' with 0, then convert to numeric.
anime_df['episodes'] = anime_df['episodes'].replace('Unknown', '0') # Replace 'Unknown' with 0 for episode count as it's a numeric column
anime_df['episodes'] = pd.to_numeric(anime_df['episodes'])

# Verify that missing values have been handled and data types are correct
display(anime_df.info())
```

```python
# Combine numerical features and genre features

# Select numerical features for the recommendation system
numerical_features = anime_df[['rating', 'members', 'episodes']]

# Normalize numerical features using MinMaxScaler
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
numerical_scaled = scaler.fit_transform(numerical_features)
numerical_df_scaled = pd.DataFrame(numerical_scaled, columns=numerical_features.columns, index=anime_df.index)

# Combine scaled numerical features with genre features
feature_matrix = pd.concat([numerical_df_scaled, genre_features], axis=1)

# Display the first few rows of the combined feature matrix
display(feature_matrix.head())
```

```python
# Recommendation System: Cosine Similarity

from sklearn.metrics.pairwise import cosine_similarity

def recommend_anime(anime_title, feature_matrix, anime_df, num_recommendations=10):
    # Get the index of the target anime
    if anime_title not in anime_df['name'].values:
        print(f"Anime '{anime_title}' not found in the dataset.")
        return pd.DataFrame()

    idx = anime_df[anime_df['name'] == anime_title].index[0]

    # Get the feature vector for the target anime
    target_anime_features = feature_matrix.loc[idx].values.reshape(1, -1)

    # Calculate cosine similarity between the target anime and all other anime
    similarity_scores = cosine_similarity(target_anime_features, feature_matrix)

    # Get the indices of anime sorted by similarity scores (descending)
    # Flatten the similarity_scores array and get sorted indices
    sorted_indices = similarity_scores.flatten().argsort()[::-1]

    # Exclude the target anime itself from recommendations
    recommended_anime_indices = [i for i in sorted_indices if i != idx][:num_recommendations]

    # Get the recommended anime details
    recommendations = anime_df.loc[recommended_anime_indices, ['name', 'genre', 'type', 'rating', 'members']]
    recommendations['similarity_score'] = similarity_scores.flatten()[recommended_anime_indices]

    return recommendations

# Example Usage:
# Let's recommend anime similar to 'Kimi no Na wa.'
target_anime = 'Kimi no Na wa.'
recommendations_df = recommend_anime(target_anime, feature_matrix, anime_df)

display(f"Recommendations for '{target_anime}':")
display(recommendations_df)
```

```python
# Experiment with different threshold values for similarity scores

# Modify the recommend_anime function to include a similarity threshold
def recommend_anime_with_threshold(anime_title, feature_matrix, anime_df, num_recommendations=10, similarity_threshold=0.5):
    if anime_title not in anime_df['name'].values:
        print(f"Anime '{anime_title}' not found in the dataset.")
        return pd.DataFrame()

    idx = anime_df[anime_df['name'] == anime_title].index[0]
    target_anime_features = feature_matrix.loc[idx].values.reshape(1, -1)
    similarity_scores = cosine_similarity(target_anime_features, feature_matrix).flatten()

    # Filter recommendations based on similarity threshold and exclude the target anime itself
    eligible_indices = [i for i, score in enumerate(similarity_scores) if score >= similarity_threshold and i != idx]

    # Sort eligible indices by similarity score in descending order
    sorted_eligible_indices = sorted(eligible_indices, key=lambda x: similarity_scores[x], reverse=True)

    # Get the top N recommendations
    recommended_anime_indices = sorted_eligible_indices[:num_recommendations]

    if not recommended_anime_indices:
        print(f"No recommendations found for '{anime_title}' with similarity threshold {similarity_threshold}.")
        return pd.DataFrame()

    recommendations = anime_df.loc[recommended_anime_indices, ['name', 'genre', 'type', 'rating', 'members']]
    recommendations['similarity_score'] = similarity_scores[recommended_anime_indices]

    return recommendations

# Example Usage with a threshold:
# Let's try with a higher similarity threshold and fewer recommendations
target_anime = 'Kimi no Na wa.'
recommendations_high_threshold = recommend_anime_with_threshold(target_anime, feature_matrix, anime_df, num_recommendations=5, similarity_threshold=0.9)

display(f"Recommendations for '{target_anime}' with similarity threshold 0.9 (Top 5):")
display(recommendations_high_threshold)

# Example Usage with a lower similarity threshold and more recommendations
recommendations_low_threshold = recommend_anime_with_threshold(target_anime, feature_matrix, anime_df, num_recommendations=15, similarity_threshold=0.8)

display(f"Recommendations for '{target_anime}' with similarity threshold 0.8 (Top 15):")
display(recommendations_low_threshold)
```

```python
# Feature Extraction: Genre

# Split the 'genre' column into individual genres and create a set of all unique genres
all_genres = set()
for genres_list in anime_df['genre'].str.split(', '):
    for genre in genres_list:
        if genre:
            all_genres.add(genre.strip())

# Create a new DataFrame for genre features (one-hot encoding)
genre_features = pd.DataFrame(0, index=anime_df.index, columns=list(all_genres))

# Populate the genre_features DataFrame
for index, row in anime_df.iterrows():
    genres = row['genre'].split(', ')
    for genre in genres:
        if genre.strip() in genre_features.columns:
            genre_features.loc[index, genre.strip()] = 1

# Display the first few rows of the genre_features DataFrame
display(genre_features.head())
```
