```python
import scipy.stats as stats
import math

print("=" * 60)
print("  HYPOTHESIS TESTING - BOMBAY HOSPITALITY LTD.")
print("=" * 60)

# ─────────────────────────────────────────────
# Given Data
# ─────────────────────────────────────────────
X_mean      = 600       # Mean units produced per week
X_std       = 25        # Std deviation of units produced
sample_mean = 3050      # Observed sample mean weekly cost (Rs.)
n           = 25        # Sample size

# Theoretical mean cost from the model: W = 1000 + 5X
mu = 1000 + 5 * X_mean  # = 1000 + 3000 = 4000... wait, let's check
# The problem states μ = 1000 + 5*600
theoretical_mean = 1000 + 5 * X_mean
print(f"\nTheoretical mean (μ) = 1000 + 5×{X_mean} = Rs. {theoretical_mean}")

# Population std deviation of W:  σ_W = 5 * σ_X  (since W = 1000 + 5X)
sigma = 5 * X_std       # = 125
print(f"Population std dev (σ) = 5 × {X_std} = {sigma}")

alpha = 0.05

# ─────────────────────────────────────────────
# STEP 1: State the Hypotheses
# ─────────────────────────────────────────────
print("\n" + "─" * 60)
print("STEP 1: Hypotheses")
print("─" * 60)
print("H₀ (Null Hypothesis)     : μ = 4000  (costs follow the model)")
print("H₁ (Alternate Hypothesis): μ > 4000  (costs are HIGHER than model)")
print("→ This is a ONE-TAILED (right-tailed) test.")

# ─────────────────────────────────────────────
# STEP 2: Calculate the Test Statistic (z)
# ─────────────────────────────────────────────
print("\n" + "─" * 60)
print("STEP 2: Test Statistic")
print("─" * 60)

# Formula: z = (x̄ - μ) / (σ / √n)
std_error = sigma / math.sqrt(n)
z_statistic = (sample_mean - theoretical_mean) / std_error

print(f"  x̄ (sample mean)    = {sample_mean}")
print(f"  μ (theoretical)    = {theoretical_mean}")
print(f"  σ (pop. std dev)   = {sigma}")
print(f"  n (sample size)    = {n}")
print(f"  Standard Error     = σ/√n = {sigma}/√{n} = {std_error:.4f}")
print(f"\n  z = (x̄ - μ) / SE")
print(f"  z = ({sample_mean} - {theoretical_mean}) / {std_error:.4f}")
print(f"  z = {sample_mean - theoretical_mean} / {std_error:.4f}")
print(f"  z = {z_statistic:.4f}")

# ─────────────────────────────────────────────
# STEP 3: Determine the Critical Value & p-value
# ─────────────────────────────────────────────
print("\n" + "─" * 60)
print("STEP 3: Critical Value & p-value  (α = 0.05, right-tailed)")
print("─" * 60)

critical_value = stats.norm.ppf(1 - alpha)   # z_critical for right tail
p_value        = 1 - stats.norm.cdf(z_statistic)

print(f"  Critical value (z_α)  = {critical_value:.4f}")
print(f"  p-value               = {p_value:.6f}")
print(f"  α                     = {alpha}")

# ─────────────────────────────────────────────
# STEP 4: Decision
# ─────────────────────────────────────────────
print("\n" + "─" * 60)
print("STEP 4: Decision")
print("─" * 60)

if z_statistic > critical_value:
    decision = "REJECT H₀"
    reject = True
else:
    decision = "FAIL TO REJECT H₀"
    reject = False

print(f"  z-statistic ({z_statistic:.4f})  vs  critical value ({critical_value:.4f})")
print(f"  p-value     ({p_value:.6f}) vs  α ({alpha})")
print(f"\n  Decision: {decision}")

# ─────────────────────────────────────────────
# STEP 5: Conclusion
# ─────────────────────────────────────────────
print("\n" + "─" * 60)
print("STEP 5: Conclusion")
print("─" * 60)

if reject:
    print("""  There IS sufficient statistical evidence at the 5% significance
  level to support the restaurant owners' claim that the weekly
  operating costs are HIGHER than what the model W = $1,000 + $5X
  predicts.

  The observed mean cost of Rs. 3,050 (with n=25 restaurants) leads
  to a z-score that falls in the critical rejection region, indicating
  the difference is statistically significant and not due to chance.""")
else:
    print("""  There is NOT sufficient statistical evidence at the 5% significance
  level to support the restaurant owners' claim. The weekly operating
  costs are consistent with the model W = $1,000 + $5X.""")

print("\n" + "=" * 60)
```

Key numbers from  assignment:

Theoretical mean: μ = 1000 + 5×600 = Rs. 4,000
Population σ = 5 × 25 = 125
Standard error = 125/√25 = 25
z-statistic = (3050 − 4000) / 25 = −38

Interesting finding: The sample mean of Rs. 3,050 is actually lower than the theoretical mean of Rs. 4,000 — so the z-score comes out deeply negative (−38). Since this is a right-tailed test (owners claim costs are higher), we fail to reject H₀. The data does not support the claim that costs are higher.

```python
import numpy as np
import scipy.stats as stats

print("=" * 65)
print("  CONFIDENCE INTERVAL - PRINT-HEAD DURABILITY ANALYSIS")
print("=" * 65)

# ─────────────────────────────────────────────
# Data
# ─────────────────────────────────────────────
data = [1.13, 1.55, 1.43, 0.92, 1.25, 1.36, 1.32,
        0.85, 1.07, 1.48, 1.20, 1.33, 1.18, 1.22, 1.29]

n           = len(data)
x_bar       = np.mean(data)
s           = np.std(data, ddof=1)   # sample std deviation (ddof=1)
sigma_known = 0.2                    # known population std deviation (part b)
confidence  = 0.99
alpha       = 1 - confidence

print(f"\nSample data (n = {n}):")
print(" ", data)
print(f"\n  Sample mean (x̄)              = {x_bar:.4f} million characters")
print(f"  Sample std deviation (s)     = {s:.4f} million characters")
print(f"  Known population σ           = {sigma_known}")
print(f"  Confidence level             = {int(confidence*100)}%")
print(f"  α                            = {alpha}")

# ═══════════════════════════════════════════════════════════════
# PART A: 99% CI using SAMPLE standard deviation  → t-distribution
# ═══════════════════════════════════════════════════════════════
print("\n" + "─" * 65)
print("PART A: 99% CI Using Sample Standard Deviation (t-distribution)")
print("─" * 65)

print("""
  Why t-distribution?
  • The population standard deviation (σ) is UNKNOWN.
  • Sample size is small (n = 15 < 30).
  • We use the sample std deviation (s) as an estimate of σ.
  • Under these conditions the t-distribution (with n-1 degrees
    of freedom) gives more accurate critical values than z,
    accounting for the extra uncertainty from estimating σ.
""")

df        = n - 1                                       # degrees of freedom
t_critical = stats.t.ppf(1 - alpha / 2, df=df)         # two-tailed critical value
margin_a  = t_critical * (s / np.sqrt(n))
lower_a   = x_bar - margin_a
upper_a   = x_bar + margin_a

print(f"  Steps:")
print(f"  1. Degrees of freedom  = n - 1 = {n} - 1 = {df}")
print(f"  2. t critical value (α/2 = {alpha/2}, df = {df})")
print(f"                         t = {t_critical:.4f}")
print(f"  3. Standard error      = s / √n = {s:.4f} / √{n} = {s/np.sqrt(n):.4f}")
print(f"  4. Margin of error     = t × SE = {t_critical:.4f} × {s/np.sqrt(n):.4f} = {margin_a:.4f}")
print(f"\n  CI Formula: x̄ ± t × (s / √n)")
print(f"            = {x_bar:.4f} ± {margin_a:.4f}")
print(f"\n  ✦ 99% Confidence Interval (t): ({lower_a:.4f}, {upper_a:.4f})")
print(f"    i.e., ({lower_a:.2f}M, {upper_a:.2f}M) million characters")

# ═══════════════════════════════════════════════════════════════
# PART B: 99% CI using KNOWN population standard deviation → z
# ═══════════════════════════════════════════════════════════════
print("\n" + "─" * 65)
print("PART B: 99% CI Using Known Population Std Deviation (z-distribution)")
print("─" * 65)

print("""
  Why z-distribution?
  • The population standard deviation (σ = 0.2) is KNOWN.
  • When σ is known we use the standard normal (z) distribution
    directly — no need to estimate σ from the sample.
""")

z_critical = stats.norm.ppf(1 - alpha / 2)             # two-tailed z critical value
margin_b   = z_critical * (sigma_known / np.sqrt(n))
lower_b    = x_bar - margin_b
upper_b    = x_bar + margin_b

print(f"  Steps:")
print(f"  1. z critical value (α/2 = {alpha/2})")
print(f"                         z = {z_critical:.4f}")
print(f"  2. Standard error      = σ / √n = {sigma_known} / √{n} = {sigma_known/np.sqrt(n):.4f}")
print(f"  3. Margin of error     = z × SE = {z_critical:.4f} × {sigma_known/np.sqrt(n):.4f} = {margin_b:.4f}")
print(f"\n  CI Formula: x̄ ± z × (σ / √n)")
print(f"            = {x_bar:.4f} ± {margin_b:.4f}")
print(f"\n  ✦ 99% Confidence Interval (z): ({lower_b:.4f}, {upper_b:.4f})")
print(f"    i.e., ({lower_b:.2f}M, {upper_b:.2f}M) million characters")

# ─────────────────────────────────────────────
# Comparison Summary
# ─────────────────────────────────────────────
print("\n" + "─" * 65)
print("SUMMARY COMPARISON")
print("─" * 65)
print(f"  {'Method':<35} {'Lower':>8}  {'Upper':>8}  {'Width':>8}")
print(f"  {'─'*35} {'─'*8}  {'─'*8}  {'─'*8}")
print(f"  {'Part A — t (unknown σ)':<35} {lower_a:>8.4f}  {upper_a:>8.4f}  {upper_a-lower_a:>8.4f}")
print(f"  {'Part B — z (known σ=0.2)':<35} {lower_b:>8.4f}  {upper_b:>8.4f}  {upper_b-lower_b:>8.4f}")
print(f"""
  Observation:
  The t-interval (Part A) is WIDER than the z-interval (Part B).
  This is expected — using an estimated s instead of the known σ
  introduces additional uncertainty, which the t-distribution
  compensates for by producing a larger critical value
  (t = {t_critical:.4f} vs z = {z_critical:.4f}).
""")
print("=" * 65)
```
